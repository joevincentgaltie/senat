```python
newdf=pd.DataFrame()
newdf['Sénateur']=0
newdf['Groupe d étude']=0
newdf['Poste']=0 
for i in range(21) : 
    newdf.loc[i,'Sénateur']=groupes_etudes.iloc[i,1]
    newdf.loc[i,'Poste']='Président'
    newdf.loc[i,'Groupe d étude']=groupes_etudes.iloc[i,0]
for i in range(21,42) : 
    newdf.loc[i,'Sénateur']=groupes_etudes.iloc[i-21,2]
    newdf.loc[i,'Poste']='Vice-Président'
    newdf.loc[i,'Groupe d étude']=groupes_etudes.iloc[i-21,0]
```
